```python
from tools import *
#print(json.dumps(str(globals()), indent=2))

def call_function_from_response(response):
    # Parse the response assuming it's a JSON string
    results = []
    for response_data in response:
        # Iterate over each function call in the response
        func_call = response_data.get('function_call', {})
        func_name = func_call.get("name")
        func_args = func_call.get("parameters", {})
        if func_name and func_name in globals():
            function_to_call = globals()[func_name]
            result = None

            # Inspect the function to get its parameters
            params = inspect.signature(function_to_call).parameters

            # Call the function with its arguments
            if len(params) == 1:
                # Single argument, unpack the dict and pass the value
                arg_name = next(iter(params))
                x = {}
                x["function_call"] = {}
                x["function_call"]["name"] = func_name
                x["function_call"]["parameters"] = func_args
                x["function_call"]["result"] = function_to_call(func_args[arg_name])
                result = x["function_call"]["result"]
                results.append(x)
                return results
            else:
                # Multiple arguments, pass the dict as kwargs
                x = {}
                x["function_call"] = {}
                x["function_call"]["name"] = func_name
                x["function_call"]["parameters"] = func_args
                x["function_call"]["result"] = function_to_call(**func_args)
                result = x["function_call"]["result"]
                results.append(x)
                return results
        else:
            print(f"Function '{func_name}' is not available.")
    return results

system_content = SYSTEM_PROMPT.format(funcs=json.dumps(functions()))

prompt_template = """
<|im_start|>system
  {system_content}
<|im_end|>

<|im_start|>user
  {user_content}
<|im_end|>

<|im_start|>assistant
"""

output_chars = []
grammar = LlamaGrammar.from_file("./json_arr.gbnf")
user_content = "Current temp in SF" #llm(prompt="Generate a random prompt for an LLM agent that can search the web: ", echo=True, stream=False, temperature=0.0)["choices"][0]["text"]

# 

prompt_ = prompt_template.format(system_content=system_content, user_content=user_content)

# print(prompt_, "\n\n")

output = llm(prompt=prompt_, max_tokens=256, stop=["<|im_end|>"], echo=True, stream=True, grammar=grammar, temperature=1.0, mirostat_mode = 2, mirostat_tau=2.0, mirostat_eta=1.1)

# print(output)

count = 0

for oo in output:
    text = oo["choices"][0]["text"]
    print(text, end="")
    output_chars.append(text)
    # print(output_chars)
# output2 = llm(prompt=f"If this JSON output has any errors, can you please please make it syntactically valid? Thank you. JSON: {output}", max_tokens=1024, stop=["\n\n", "<|im_end|>"], echo=True, stream=True, mirostat_mode = 2, mirostat_tau=2.0, mirostat_eta=1.1, grammar=grammar, temperature=0.5)
# print(output2)

# for oo in output2:
#     text = oo["choices"][0]["text"]
#     print(text, end="")
#     output_chars.append(text)

# print("\n")

try:
    parsed = json.loads(''.join(output_chars))
    print(json.dumps(parsed, indent=2))
    function_outputs = call_function_from_response(parsed)
    for function_output_ in function_outputs:
        print(function_output_)
        prompt_ += "<|im_start|>assistant"+"\n"+json.dumps(function_output_["function_call"]["result"], indent=2)+"<|im_end|>"
        #print(f"prompt inside function_output_ loop: \n {prompt_}\n")

    output_chars = []
    output = llm(prompt=f"<|im_start|>system\nTry to answer the user's original question: {user_content} based on the function output: {function_outputs}<|im_end|>", max_tokens=2048, stop=["<|im_end|>"], echo=True, stream=False, mirostat_mode = 2, mirostat_tau=4.0, mirostat_eta=1.1)
    choice_text = output["choices"][0]["text"]
    print(choice_text)

except Exception as e:
    print(e)
    
# text = json.loads(output["choices"][0]["text"])
# func_name = text["function_name"]
# func_args = text["function_args"]
# print(f"{func_name}: {func_args}")

# print(json.dumps(output, indent=2))
```
